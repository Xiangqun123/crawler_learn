# pyquery

```python
from pyquery import PyQuery as pq
```

```python
import requests
doc = requests.get('https://book.douban.com').text
print(doc)
```

```python
pqdoc = pq(doc)
lists = []
items = pqdoc('#content > div > div.article > div.section.books-express > div.bd > div > div > ul:nth-child(2) > li').items()
for li in items:
#     print(li('div.cover a').attr.href)
#     print(li('div.more-meta h4.title').text())
#     print(li('div.more-meta p span.author').text())
#     print(li('div.more-meta p span.year').text())
#     print(li('div.more-meta p span.publisher').text())
    lists.append([li('div.cover a').attr.href, li('div.more-meta h4.title').text(), li('div.more-meta p span.author').text(), li('div.more-meta p span.year').text(), li('div.more-meta p span.publisher').text()])

lists
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(lists, columns = ('href', 'title', 'author', 'pub_date', 'publisher'))
```

```python
df
```

```python
df.to_csv('pyquery_csv', sep=',', index=0)
```
