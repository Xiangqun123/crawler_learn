# 正则表达式

```python
# 导入re包
import re
import requests
```

```python
content = requests.get("https://www.hcittech.cn/").text
print(content)
```

```python
pattern = re.compile("test_desc.*?href=\"#\">(.*?)</a>.*?<p>(.*?)<a href", re.S)
result = re.findall(pattern, content)
print(result)
for name, work in result:
    print(name, ":", work)
```
